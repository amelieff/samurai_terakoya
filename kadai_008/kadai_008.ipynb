```python
import random

var = random.randint(1,100)

# var の値を確認
print("変数var の値")
print(var)

print("出力")

# 3の倍数なら Fizz出力
if var % 3 == 0:
    print("Fizz")

    # さらに 5の倍数なら Buzz出力。ならびに 3の倍数かつ5の倍数であることは確定なので FizzBuzz出力。
    if var % 5 == 0:
        print("Buzz")
        print("FizzBuzz")

# 3の倍数ではなくて 5の倍数なら Buzz出力
elif var % 5 == 0:
    print("Buzz")

# 3の倍数でも5の倍数でもない場合には変数 var の値を出力
else:
    print(var)
```
