```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0, linewidth=2.0)

x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0, 1].plot(x1, y1, linewidth=2.0)
axes[0, 1].set(
    xlim=(0, 10), xticks=np.arange(0, 10), ylim=(-1, 5), yticks=np.arange(-1, 6)
)

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x2 = np.random.normal(15, 5, 2000)
plt.xlim(-10, 40)
# ヒストグラムをプロット
axes[0, 2].hist(x2, bins=30)

# モーターから観測した音の周波数
hertz = [
    300,
    200,
    200,
    500,
    600,
    2000,
    1600,
    1800,
    1850,
    1700,
    1500,
    1500,
    1500,
    1400,
    5000,
    6000,
    2000,
    3000,
    3000,
]
# 箱ひげ図
axes[1, 0].boxplot(hertz)


plt.show()
```
