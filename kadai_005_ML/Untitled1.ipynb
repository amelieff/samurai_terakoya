```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

y = [1, 2, -5, 2]
axes[0,0].plot(y)

x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)
axes[0,1].plot(x, y)

x = np.random.normal(15, 5, 2000)
axes[0,2].hist(x)

x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))
axes[1,0].scatter(x, y)

plt.show()
```
