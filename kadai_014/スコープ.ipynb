```python
# 正しい例: taxをローカル変数ではなくグローバル変数として定義
price1 = 100
price2 = 200
tax = 1.1

def total():
    return price1 + price2

print (total() * tax)
```

```python
# 正しい例2: price1, price2をtotal関数の引数として定義
price1 = 100
price2 = 200
tax = 1.1

def total(local_price1, local_price2):
    return local_price1 + local_price2

print (total(price1, price2) * tax)
```
