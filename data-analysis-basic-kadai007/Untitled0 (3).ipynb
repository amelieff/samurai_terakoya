```python
#ランダムな整数を利用するため、記述が必要です。
import random

# 変数numに0~4までのランダムな整数を代入する
num = random.randint(0, 4)

# カウンタ変数を用意する
i = 1

# 変数numの最初の値を出力する（確認用）
print(f"最初の値は{num}です。")

# 変数numの値が0以外である間、変数numの値を出力し続ける
while num != 0:
    # 変数numに0~4までのランダムな整数を代入する
    num = random.randint(0, 4)

    # カウンタ変数iの値が5であれば、break文で繰り返し処理を強制終了する
    if i == 5:
        print("5回目なので繰り返し処理を強制終了します。")
        break

    # 次の条件式で比較される、変数numの現在の値を出力する
    print(f"現在の値は{num}です。")
    #　カウンタ変数のの値を1増やす
    i = i + 1
```

```python
#ランダムな整数を利用するため、記述が必要です。
import random

# 合計用の変数を用意する
sum = 0

# 変数sumの値が20以上になるまで繰り返し処理を行う
while sum < 20:
    # 変数numに1~10までのランダムな整数を代入する
    num = random.randint(1, 10)

    print(f"{num}が出ました。")

    # 変数numの値が偶数（2で割った余りが0）であれば、加算せずにcontinue文で次のループに進む
    if num % 2 == 0:
        print("偶数なので加算しません。")
        continue

    # 変数sumに変数numの値を加算する
    sum = sum + num
    print(f"現在の合計は{sum}です。")
```

```python
def calculate_total_price(price: float, tax_rate: float) -> float:
    # priceとtax_rateからtotalを計算
    # 例: 110 * (1 + 10 / 100) = 121.0
    total = price * (1 + tax_rate / 100)
    return total

# --- 関数の実行 ---

# 第1引数に商品の金額、第2引数に消費税率を設定
product_price = 110
tax = 10

# 関数を呼び出して計算結果を受け取る
final_price = calculate_total_price(product_price, tax)

# 結果を出力
print(f"商品の金額: {product_price}円")
print(f"消費税率: {tax}%")
print(f"税込み価格: {final_price}円")
```

```python
class Human:
    """
    名前(name)と年齢(age)の属性を持つ、人間を表すクラス。
    """
    def __init__(self, name: str, age: int):
        """
        Humanオブジェクトを初期化（コンストラクタ）。

        Args:
            name (str): 人物の名前
            age (int): 人物の年齢
        """
        self.name = name
        self.age = age

    def printinfo(self):
        """
        インスタンスが持つ名前と年齢の情報を標準出力に表示するメソッド。
        """
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}")

# --- ここから実行部分 ---

# Humanクラスのインスタンスを作成し、変数に代入する
person1 = Human("田中 太郎", 30)
person2 = Human("佐藤 花子", 25)

# 各インスタンスのメソッドを実行して情報を出力する
print("--- 1人目の情報 ---")
person1.printinfo()

print("\n--- 2人目の情報 ---")
person2.printinfo()
```

```python
# datetimeモジュールをインポート
import datetime

# 現在時刻を取得する
now = datetime.datetime.now()

# 1年間の日数差を変数td_1yに代入する
td_1y = datetime.timedelta(days=365)

# 現在の日時に1年を加算し、変数addに代入する
add = now + td_1y

# 3日間の日数差を変数td_3dに代入する
td_3d = datetime.timedelta(days=3)

# 現在の日時から3日を減算し、変数subに代入する（sub＝「減算」を意味するsubtractionの略）
sub = now - td_3d

# 加算・減算した日時を特定のフォーマットで出力する
print(add.strftime("現在から1年後は%Y年%m月%d日%H時%M分%S秒です。"))
print(sub.strftime("現在から3日前は%Y年%m月%d日%H時%M分%S秒です。"))
```

```python
class Human:
    """
    名前(name)と年齢(age)の属性を持つ、人間を表すクラス。
    """
    def __init__(self, name: str, age: int):
        """
        Humanオブジェクトを初期化（コンストラクタ）。

        Args:
            name (str): 人物の名前
            age (int): 人物の年齢
        """
        self.name = name
        self.age = age

    def check_adult(self):
        """
        インスタンスの年齢をチェックし、大人かどうかを標準出力に表示するメソッド。
        20歳以上を大人と判定する。
        """
        if self.age >= 20:
            print(f"{self.name}さんは大人です。")
        else:
            print(f"{self.name}さんは大人ではありません。")

# --- ここから実行部分 ---

# Humanクラスのインスタンスを複数生成
person1 = Human("遠藤 遼一", 25)
person2 = Human("森岡 賢", 18)
person3 = Human("藤井 麻輝", 32)
person4 = Human("上領 亘", 15)

# 生成したインスタンスをリストに追加
people_list = [person1, person2, person3, person4]

# リストの要素を一つずつ取り出し、メソッドを実行する
for person in people_list:
    person.check_adult()
```

```python
import random

START_POSITION = 0  # スタートのマス目
GOAL_POSITION = 10  # ゴールのマス目

now_pos = START_POSITION  # 現在位置
roll_count = 0            # 振った回数
items = []                # アイテムを格納するリスト

# 3の倍数のマス目にアイテムを配置する
for i in range(1, GOAL_POSITION):  # 1~9
    if i % 3 == 0:
        items.append("アイテム")
    else:
        items.append("")

print("すごろくゲームを開始します！")

# ゴールにたどり着くまでサイコロを振り続ける
while now_pos != GOAL_POSITION:
    # サイコロを振る
    dice = random.randint(1, 6)

    # 現在位置を進める
    now_pos += dice
    # 振った回数を加算
    roll_count += 1

    print(f"{roll_count}回目：{dice}が出ました！現在{now_pos}マス目です。")

    if now_pos > GOAL_POSITION:  # ゴールを超えた場合は振り出しに戻る
        now_pos = START_POSITION
        print("ゴールを超えたため振り出しに戻ります。")
    elif now_pos == GOAL_POSITION:  # ゴールにピッタリ止まったらクリア
        print("ゴールに到達しました！")
    elif items[now_pos] != "":  # アイテムがあれば取得
        print("アイテムを獲得しました！")

print("おめでとうございます！ゲームクリアです。")
```

```python
# STEP1
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

df = pd.read_csv("sample_pandas_6.csv")

category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# STEP2
category_counts = df['カテゴリー'].value_counts()
category_counts.plot(kind='bar')

# STEP3
plt.show()

# STEP4
# --- 商品番号ごとの注文数を集計する (合計値の例) ---
# df.groupby('基準列')['サブ列'].集計関数() の形式
order_quantity_by_product_sum = df.groupby('商品番号')['注文数'].sum()
print("--- 商品番号ごとの注文数合計 ---")
print(order_quantity_by_product_sum.head())

# --- 商品番号ごとの注文数の基本統計量を算出する ---
# df.groupby('基準列')['サブ列'].describe() の形式
order_quantity_by_product_description = df.groupby('商品番号')['注文数'].describe()
print("\n--- 商品番号ごとの注文数の基本統計量 ---")
print(order_quantity_by_product_description.head())
```
