```python
import matplotlib.pyplot as plt
import numpy as np

# データの準備
x = np.linspace(0, 10, 100)
y = x**2
sin_y = np.sin(x)
data = np.random.randn(1000)
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

# キャンバスの準備（2行×3列）
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

# 1. 折れ線グラフ
axes[0].plot(x, y, label="y = x^2", color='blue')
axes[0].set_title("Line Plot")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].legend()

# 2. sin関数
axes[1].plot(x, sin_y, label="y = sin(x)", color='green')
axes[1].set_title("Sine Function")
axes[1].set_xlabel("x")
axes[1].set_ylabel("sin(x)")
axes[1].legend()

# 3. ヒストグラム
axes[2].hist(data, bins=30, color='orange')
axes[2].set_title("Histogram")
axes[2].set_xlabel("Value")
axes[2].set_ylabel("Frequency")

# 4. 散布図
axes[3].scatter(x_scatter, y_scatter, color='red')
axes[3].set_title("Scatter Plot")
axes[3].set_xlabel("x")
axes[3].set_ylabel("y")

# 5, 6番目のグラフは空白に
for i in [4, 5]:
    axes[i].axis('off')
    axes[i].set_title("Empty")

# レイアウト調整
plt.tight_layout()
plt.show()
```
