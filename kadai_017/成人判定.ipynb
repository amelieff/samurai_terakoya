```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}歳")

    # 大人かどうかを判定して出力するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}（{self.age}歳）は大人です。")
        else:
            print(f"{self.name}（{self.age}歳）は大人ではありません。")
```

```python
people_list = [
    Human("田中太郎", 30),
    Human("山田花子", 18),
    Human("鈴木一郎", 14),
    Human("佐藤愛", 29)
]


for person in people_list:
    person.check_adult()
```
