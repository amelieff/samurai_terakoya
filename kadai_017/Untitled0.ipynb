```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print("adult")
    else:
      print("child")

user1 = Human("hoge", 0)
user2 = Human("huga", 20)
user3 = Human("piyo", 100)

users = [user1, user2, user3]
for i in users:
  i.check_adult()
```
