```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f'{self.name}は大人である')
    else:
      print(f'{self.name}は大人でない')

array_ins = []
array_ins.append(Human('tanaka', 20))
array_ins.append(Human('suzuki', 21))
array_ins.append(Human('takahashi', 19))
array_ins.append(Human('yamamoto', 34))

for ins in array_ins:
  ins.check_adult()
```
