```python
class Human:
  def __init__(self):
    self.name = ""
    self.age = 0

  def check_adult(self):
    if self.age >= 20:
      print(f'{self.name}は大人である')
    else:
      print(f'{self.name}は大人でない')

array_human = [{'name':'tanaka', 'age':20},{'name':'suzuki', 'age':21},{'name':'takahashi', 'age':19},{'name':'yamamoto', 'age':34} ]
array_ins = []
for human in array_human:
  human_ins = Human()
  human_ins.name = human['name']
  human_ins.age = human['age']
  array_ins.append(human_ins)

for ins in array_ins:
  ins.check_adult()
```
