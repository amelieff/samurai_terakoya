```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def printinfo(self):
    print(f"Name: {self.name}, Age: {self.age}")

  def check_adalt(self):
    if self.age >= 20:
      print("Adult")
    else:
      print("Child")

humanA = Human("Yukinobu", 42)
humanB = Human("Eri", 19)
humans = [humanA, humanB]

for human in humans:
  human.check_adalt()
```
