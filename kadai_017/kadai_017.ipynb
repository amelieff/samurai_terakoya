```python
class Human:
    def __init__(self):
        self.name = ""
        self.age = ""

    def set_name_age(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            return "大人です"
        else:
            return "大人ではありません"

    def show_name_age(self):
        return self.name + "さんは、" + str(self.age) + "歳なので"

taro = Human()
taro.set_name_age("太郎", 20)

hanako = Human()
hanako.set_name_age("花子", 15)

jiro = Human()
jiro.set_name_age("ジロー", 31)

member = [taro, hanako, jiro]

for i in member:
    print(i.show_name_age(), i.check_adult())
```
