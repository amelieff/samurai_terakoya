```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print("大人です")
    else:
      print("大人ではありません")

user = []
users = {"太郎":40, "次郎":30,"三郎":20,"四郎":10 }
for name in users:
  user.append(Human(name, users[str(name)]))

for i_user in user:
  print(str(i_user.name) + ":" + str(i_user.age)+ ":" )
  i_user.check_adult()
```
