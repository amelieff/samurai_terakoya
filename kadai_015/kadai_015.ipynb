```python
class Human:
    def __init__(self):
        self.name = ""
        self.age = ""

    def set_name_age(self, name, age):
        self.name = name
        self.age = age

    def show_name_age(self):
        print(self.name)
        print(self.age)

taro = Human()

taro.set_name_age("太郎", 20)

taro.show_name_age()
```
