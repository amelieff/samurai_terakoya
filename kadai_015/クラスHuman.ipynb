```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}歳")
```

```python
person = Human("田中太郎", 30)
person.printinfo()
```

```python
person = Human("山田花子", 18)
person.printinfo()
```
